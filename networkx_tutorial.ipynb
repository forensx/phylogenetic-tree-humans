```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
G = nx.Graph()

# adding nodes
G.add_node("Berea College")
G.add_nodes_from(range(1, 10))
```

```python
G.remove_node("Berea College")
```

```python
G.add_edge(5, 100)
G.remove_edge(5, 100)
#G.add_edges_from()
```

```python
# we can use weighted edges
G.add_weighted_edges_from([(0,1, 1), (1, 4, 7.5)])
```

```python
G.add_path([1,2,3])
```

```python
nx.draw(G, with_labels = True)
plt.draw()
plt.show()
```
